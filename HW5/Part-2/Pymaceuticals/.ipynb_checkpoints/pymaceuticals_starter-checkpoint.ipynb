```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv(mouse_drug_data_to_load)
clinical_df = pd.read_csv(clinical_trial_data_to_load)

clinical_df.head()
# Combine the data into a single dataset
df = pd.DataFrame.merge(mouse_df, clinical_df, on='Mouse ID')
# Display the data table for preview

df
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
grouped = df.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].mean()

grouped
# Convert to DataFrame
group_df = pd.DataFrame(grouped)

# Preview DataFrame
group_df.head()
```

```python
# Generate the Plot (with Error Bars)
#plt.plot(group_df.index(), group_df['Tumor Volume (mm3)'])
# Save the Figure
gd = group_df.unstack().transpose()
#gd = gd.reset_index()
#gd = gd.drop('level_0', 1)
gd = gd.reset_index(level=0, drop=True)
gd
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.plot(gd.index, gd['Capomulin'], c='r', linestyle='dashed', marker='D')
ax1.plot(gd.index, gd['Infubinol'], c='blue', linestyle='dashed', marker='^')
ax1.plot(gd.index, gd['Ketapril'], c='g', linestyle='dashed', marker='s')
ax1.plot(gd.index, gd['Placebo'], c='black', linestyle='dashed', marker='d')
plt.xlim(0, 50)
plt.ylim(30, 80)
```

```python
# Dependencies
import pandas as pd
```

```python
# load CSV
file = 'Resources/2016-FCC-New-Coders-Survey-Data.csv'
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
df = pd.read_csv(file, low_memory=True)
df.head()
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
df = df.iloc[:, [0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
df.head()
```

```python
# Change reading 0 and 1 to No and Yes, respectively
df = df.replace({0:'No', 1:'Yes'})
df.head()
```

```python
# Extract rows for only those who attended a bootcamp
df = df.loc[df['AttendedBootcamp'] == 'Yes']
df.head()
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
v = df['BootcampName'].value_counts()
dfbc = pd.DataFrame(v)
dfbc = dfbc.reset_index()
dfbc.columns = ['BootcampName', 'Attendees']
dfbc.head()
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
dfy = df.loc[df['BootcampRecommend']=='Yes']
v = dfy['BootcampName'].value_counts()
dfbc2 = pd.DataFrame(v)
dfbc2 = dfbc2.reset_index()
dfbc2.columns = ['BootcampName', 'Recommendations']
dfbc2.head()
```

```python
# Merge the two created data frames on the name of tbe bootcamp
df_merged = pd.merge(dfbc, dfbc2, on = 'BootcampName')
df_merged.head()
```

```python
# Calculate percentage of eac bootcamp's students who are recommenders
df_merged['Percent Recommended'] = round(100 * df_merged['Recommendations']/df_merged['Attendees'], 2)
df_merged.head()
```

```python
# Export to excel and remove index
df_merged.to_excel('output/output.xlsx', index=False, header=True)
```

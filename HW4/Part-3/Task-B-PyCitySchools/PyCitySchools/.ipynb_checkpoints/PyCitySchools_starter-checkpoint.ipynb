```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
school_data_complete['type'].value_counts()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
school_count = school_data_complete['school_name'].nunique()
student_count = school_data_complete['student_name'].nunique()
total_budget = school_data_complete['budget'].sum()
math_scores = school_data_complete['math_score'].mean()
english_scores = school_data_complete['reading_score'].mean()
avg_score = ((school_data_complete['math_score'] + school_data_complete['math_score'])/2).mean()
passed_math = round(100 * len(school_data_complete[school_data_complete['math_score'] >= 70])/len(school_data_complete), 2)
passed_english = round(100 * len(school_data_complete[school_data_complete['reading_score'] >= 70])/len(school_data_complete), 2)

school_summary = pd.DataFrame({'School Count':[school_count], 'Student Count':student_count, 'Total Budget':total_budget, 'Average Math Score':math_scores, 'Average English Score':english_scores, 'Average Passing Grade':avg_score, 'Percent Passing Math':passed_math, 'Percent Passing English':passed_english })
school_summary.head()
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
groups = school_data_complete.groupby('school_name')
size = groups['student_name'].nunique()
budget = groups['budget'].mean()
stype = groups['type'].unique()
bps = round(budget/size, 2)
math_s = groups['math_score'].mean()
reading_s = groups['reading_score'].mean()

math_df = school_data_complete[school_data_complete['math_score'] >= 70]
math_df_group = math_df.groupby('school_name')
math_70_count = math_df_group['math_score'].count()
math_per = round(100 * math_70_count/size, 2)

reading_df = school_data_complete[school_data_complete['reading_score'] >= 70]
reading_df_group = reading_df.groupby('school_name')
reading_70_count = reading_df_group['reading_score'].count()
reading_per = round(100 * reading_70_count/size, 2)

passing = round(math_per + reading_per)/2

top_perf = pd.DataFrame({'School Type':stype.str.get(0),
            'Total Students':size,
            'Total School Budget':budget,
            'Per Student Budget':bps,
            'Average Math Score':math_s,
            'Average Reading Score':reading_s,
            '% Passing Math':math_per,
            '% Passing Reading':reading_per,
            'Overall Passing Rate':passing })

top_perf = top_perf.sort_values(by='Overall Passing Rate', ascending=False)
top_perf.head()
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
bottom_perf = top_perf.sort_values(by='Overall Passing Rate', ascending=True)
bottom_perf.head()
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
ninth = school_data_complete[school_data_complete['grade'] == '9th']
group9 = ninth.groupby('school_name')
mean9 = round(group9['math_score'].mean(), 2)

tenth = school_data_complete[school_data_complete['grade'] == '10th']
group10 = tenth.groupby('school_name')
mean10 = round(group10['math_score'].mean(), 2)

eleventh = school_data_complete[school_data_complete['grade'] == '11th']
group11 = eleventh.groupby('school_name')
mean11 = round(group11['math_score'].mean(), 2)

twelfth = school_data_complete[school_data_complete['grade'] == '12th']
group12 = twelfth.groupby('school_name')
mean12 = round(group12['math_score'].mean(), 2)

grade_df = pd.DataFrame({'9th':mean9, '10th':mean10, '11th':mean11, '12th':mean12})
grade_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 590, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
top_perf['Spending'] = pd.cut(top_perf['Per Student Budget'], bins=spending_bins, labels=group_names, right=False)
top_perf.head()
spending_groups = top_perf.groupby('Spending')
```

```python
avg_math = spending_groups['Average Math Score'].mean()
avg_reading = spending_groups['Average Reading Score'].mean() 
avg_passing = round((avg_math + avg_reading)/2, 2)
math_per = spending_groups['% Passing Math'].mean()
reading_per = spending_groups['% Passing Reading'].mean()


spending_summary = pd.DataFrame({'Average Math Score':avg_math,
            'Average Reading Score':avg_reading,
            'Passing Math':math_per,
            'Passing Reading':reading_per,
            'Overall Passing Rate':avg_passing })

spending_summary.head()
```
